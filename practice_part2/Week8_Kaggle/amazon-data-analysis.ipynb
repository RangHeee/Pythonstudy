# Data Cleaning

##### Checking for missing values (결측치 확인)

```python
# 각 열별로 결측치의 비율을 계산하여 반환
df.isnull().sum()/df.shape[0]*100
```

##### Dropping some columns (일부 열을 삭제)


```python
# 해당 열을 데이터프레임에서 삭제
df.drop(['product_id','user_id','user_name','review_id','img_link','product_link'], axis=1, inplace=True)
# 수정된 데이터프레임의 처음 다섯 행 출력
df.head()
```

```python
# 결측치가 있는 행을 삭제하고, 그 결과로 수정된 데이터프레임을 df에 할당
df=df.dropna()
```

##### Replacing and convert into correct data type (올바른 데이터 유형으로 교체 및 변환)

```python
# 이 코드는 데이터프레임(df)의 여러 열에 대해 문자열에서 필요 없는 문자를 제거하고, 필요한 형태로 데이터를 변환하는 과정
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)

df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['discounted_price'] = df['discounted_price'].astype(float)

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(int)

df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = df['rating_count'].astype(int)

df['rating'] = df['rating'].replace('|', np.nan)
df['rating'] = df['rating'].astype(float)

# 수정된 데이터프레임의 처음 다섯 행 출력
df.head()
```

##### Dropping missing values (결측값 삭제)

```python
# 데이터프레임(df)의 구조, 열의 데이터 타입, 결측치 유무, 그리고 메모리 사용량 등을 요약하여 출력
df.info()
```

# Creating some new features (몇 가지 새 기능 생성)

##### Splitting category column and creating four new categories
##### 카테고리 열 분할 및 4개의 새 카테고리 생성

```python
# 첫 번째 열('category')의 문자열을 '|'을 기준으로 나누어 각각의 카테고리로 분할하고, 새로운 열로 생성
df_cat = df['category'].str.split('|', expand=True).rename(columns={0:'category_1', 1:'category_2', 2:'category_3', 3:'category_4',4:'category_5'})

# 원래 데이터프레임(df)에 새로 생성된 카테고리 열들을 추가
df['category_1'] = df_cat['category_1']
df['category_2'] = df_cat['category_2']
df['category_3'] = df_cat['category_3']
df['category_4'] = df_cat['category_4']

# 기존 'category' 열을 삭제
df.drop('category', axis=1, inplace=True)

# 수정된 데이터프레임의 처음 다섯 행 출력
df.head()
```

```python
# 데이터프레임(df)에서 임의로 50개의 행을 무작위로 선택하여 출력
df.sample(50)
```

```python
# 데이터프레임(df)의 각 열별로 결측치의 개수를 계산하여 출력
df.isnull().sum()
```

```python
# 데이터프레임(df)에서 결측치가 있는 행을 삭제하고, 그 결과로 수정된 데이터프레임을 df에 할당
df=df.dropna()
```

```python
# 'discounted_price' 열과 'rating_count' 열의 값을 곱하여 'Price' 열 생성
df['Price']=df['discounted_price']*df['rating_count']

# 'Price' 열의 값을 출력
df['Price']
```

# EDA

##### Average, Minimum and Maximum Rating of category (카테고리별 평균, 최소 및 최대 평점)

```python
# 'category_3' 열을 기준으로 그룹화하고, 각 그룹에 속하는 'rating' 열의 평균(mean), 최소값(min), 최대값(max)을 계산하여 새로운 데이터프레임(result1) 생성.
result1 = df.groupby('category_3')['rating'].agg(['mean', 'min', 'max'])

# result1 데이터프레임의 'mean' 열의 값을 소수점 둘째 자리까지 반올림하여 변경.
result1['mean'] = result1['mean'].round(2)

# 수정된 result1 데이터프레임을 출력. 결과는 'category_3'별로 평균, 최소, 최대 평점이 나타남.
result1
```

##### Mean and Minimum Sell of each category (각 카테고리의 평균 및 최소 판매량)

```python
# 'category_3'를 기준으로 그룹화하고, 각 그룹에 속하는 'Price' 열의 평균(mean)과 최소값(min)을 계산하여 새로운 데이터프레임(result1)을 생성
result1 = df.groupby('category_3')['Price'].agg(['mean', 'min'])

# result1 데이터프레임의 'mean' 열의 값을 소수점 둘째 자리까지 반올림하여 변경
result1['mean'] = result1['mean'].round(2)

# 수정된 result1 데이터프레임을 출력
result1

# 이 코드는 데이터프레임(df)을 'category_3' 열을 기준으로 그룹화하고,
# 각 그룹에 속하는 'Price' 열의 평균과 최소값을 계산하여 새로운 데이터프레임(result1)을 생성
# 그리고 나서 'mean' 열의 값을 소수점 둘째 자리까지 반올림하여 변경하고,
# 최종적으로 수정된 result1 데이터프레임을 출력
```

```python
# 데이터프레임 result1의 인덱스를 재설정하고, 기존 인덱스를 버리며(inplace=True), 이를 데이터프레임에 반영
result1.reset_index(inplace=True)

# 데이터프레임 result1을 'category_3' 열을 기준으로 정리하고,
# 'Stat' 열에는 'mean'과 'min'을, 'Price' 열에는 각각의 값들을 저장하여 새로운 데이터프레임 result_melted를 생성
result_melted = result1.melt(id_vars='category_3', var_name='Stat', value_name='Price')

# Seaborn을 사용하여 각 카테고리의 평균 및 최소 판매량을 선 그래프로 시각화
# 'category_3'가 x축에, 'Price'가 y축에, 'Stat'에는 'mean'과 'min'을 구분하여 각각의 선을 표시
plt.figure(figsize=(10, 6))
sns.lineplot(x='category_3', y='Price', hue='Stat', data=result_melted, marker='o')

# 차트의 제목과 라벨을 설정
plt.title('Price Summary by Category')
plt.xlabel('Category')
plt.ylabel('Total Sell')
# '통계량'은 범례의 제목
plt.legend(title='Statistic')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# 결과는 각 카테고리의 평균 및 최소 판매량을 나타내는 선 그래프
```
