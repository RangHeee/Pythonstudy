# Data Cleaning

##### Checking for missing values (Í≤∞Ï∏°Ïπò ÌôïÏù∏)

```python
# Í∞Å Ïó¥Î≥ÑÎ°ú Í≤∞Ï∏°ÏπòÏùò ÎπÑÏú®ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Î∞òÌôò
df.isnull().sum()/df.shape[0]*100
```

##### Dropping some columns (ÏùºÎ∂Ä Ïó¥ÏùÑ ÏÇ≠Ï†ú)


```python
# Ìï¥Îãπ Ïó¥ÏùÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú ÏÇ≠Ï†ú
df.drop(['product_id','user_id','user_name','review_id','img_link','product_link'], axis=1, inplace=True)
# ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Ï≤òÏùå Îã§ÏÑØ Ìñâ Ï∂úÎ†•
df.head()
```

```python
# Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî ÌñâÏùÑ ÏÇ≠Ï†úÌïòÍ≥†, Í∑∏ Í≤∞Í≥ºÎ°ú ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ dfÏóê Ìï†Îãπ
df=df.dropna()
```

##### Replacing and convert into correct data type (Ïò¨Î∞îÎ•∏ Îç∞Ïù¥ÌÑ∞ Ïú†ÌòïÏúºÎ°ú ÍµêÏ≤¥ Î∞è Î≥ÄÌôò)

```python
# Ïù¥ ÏΩîÎìúÎäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)Ïùò Ïó¨Îü¨ Ïó¥Ïóê ÎåÄÌï¥ Î¨∏ÏûêÏó¥ÏóêÏÑú ÌïÑÏöî ÏóÜÎäî Î¨∏ÏûêÎ•º Ï†úÍ±∞ÌïòÍ≥†, ÌïÑÏöîÌïú ÌòïÌÉúÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÌôòÌïòÎäî Í≥ºÏ†ï
df['actual_price'] = df['actual_price'].str.replace('‚Çπ', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)

df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', '').str.replace(',', '')
df['discounted_price'] = df['discounted_price'].astype(float)

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(int)

df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = df['rating_count'].astype(int)

df['rating'] = df['rating'].replace('|', np.nan)
df['rating'] = df['rating'].astype(float)

# ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Ï≤òÏùå Îã§ÏÑØ Ìñâ Ï∂úÎ†•
df.head()
```

##### Dropping missing values (Í≤∞Ï∏°Í∞í ÏÇ≠Ï†ú)

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)Ïùò Íµ¨Ï°∞, Ïó¥Ïùò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ, Í≤∞Ï∏°Ïπò Ïú†Î¨¥, Í∑∏Î¶¨Í≥† Î©îÎ™®Î¶¨ ÏÇ¨Ïö©Îüâ Îì±ÏùÑ ÏöîÏïΩÌïòÏó¨ Ï∂úÎ†•
df.info()
```

# Creating some new features (Î™á Í∞ÄÏßÄ ÏÉà Í∏∞Îä• ÏÉùÏÑ±)

##### Splitting category column and creating four new categories
##### Ïπ¥ÌÖåÍ≥†Î¶¨ Ïó¥ Î∂ÑÌï† Î∞è 4Í∞úÏùò ÏÉà Ïπ¥ÌÖåÍ≥†Î¶¨ ÏÉùÏÑ±

```python
# Ï≤´ Î≤àÏß∏ Ïó¥('category')Ïùò Î¨∏ÏûêÏó¥ÏùÑ '|'ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇòÎàÑÏñ¥ Í∞ÅÍ∞ÅÏùò Ïπ¥ÌÖåÍ≥†Î¶¨Î°ú Î∂ÑÌï†ÌïòÍ≥†, ÏÉàÎ°úÏö¥ Ïó¥Î°ú ÏÉùÏÑ±
df_cat = df['category'].str.split('|', expand=True).rename(columns={0:'category_1', 1:'category_2', 2:'category_3', 3:'category_4',4:'category_5'})

# ÏõêÎûò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)Ïóê ÏÉàÎ°ú ÏÉùÏÑ±Îêú Ïπ¥ÌÖåÍ≥†Î¶¨ Ïó¥Îì§ÏùÑ Ï∂îÍ∞Ä
df['category_1'] = df_cat['category_1']
df['category_2'] = df_cat['category_2']
df['category_3'] = df_cat['category_3']
df['category_4'] = df_cat['category_4']

# Í∏∞Ï°¥ 'category' Ïó¥ÏùÑ ÏÇ≠Ï†ú
df.drop('category', axis=1, inplace=True)

# ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Ï≤òÏùå Îã§ÏÑØ Ìñâ Ï∂úÎ†•
df.head()
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)ÏóêÏÑú ÏûÑÏùòÎ°ú 50Í∞úÏùò ÌñâÏùÑ Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌïòÏó¨ Ï∂úÎ†•
df.sample(50)
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)Ïùò Í∞Å Ïó¥Î≥ÑÎ°ú Í≤∞Ï∏°ÏπòÏùò Í∞úÏàòÎ•º Í≥ÑÏÇ∞ÌïòÏó¨ Ï∂úÎ†•
df.isnull().sum()
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)ÏóêÏÑú Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî ÌñâÏùÑ ÏÇ≠Ï†úÌïòÍ≥†, Í∑∏ Í≤∞Í≥ºÎ°ú ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ dfÏóê Ìï†Îãπ
df=df.dropna()
```

```python
# 'discounted_price' Ïó¥Í≥º 'rating_count' Ïó¥Ïùò Í∞íÏùÑ Í≥±ÌïòÏó¨ 'Price' Ïó¥ ÏÉùÏÑ±
df['Price']=df['discounted_price']*df['rating_count']

# 'Price' Ïó¥Ïùò Í∞íÏùÑ Ï∂úÎ†•
df['Price']
```

##### Average, Minimum and Maximum Rating of category (Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÌèâÍ∑†, ÏµúÏÜå Î∞è ÏµúÎåÄ ÌèâÏ†ê)

```python
# 'category_3' Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôîÌïòÍ≥†, Í∞Å Í∑∏Î£πÏóê ÏÜçÌïòÎäî 'rating' Ïó¥Ïùò ÌèâÍ∑†(mean), ÏµúÏÜåÍ∞í(min), ÏµúÎåÄÍ∞í(max)ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(result1) ÏÉùÏÑ±.
result1 = df.groupby('category_3')['rating'].agg(['mean', 'min', 'max'])

# result1 Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'mean' Ïó¥Ïùò Í∞íÏùÑ ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶ºÌïòÏó¨ Î≥ÄÍ≤Ω.
result1['mean'] = result1['mean'].round(2)

# ÏàòÏ†ïÎêú result1 Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï∂úÎ†•. Í≤∞Í≥ºÎäî 'category_3'Î≥ÑÎ°ú ÌèâÍ∑†, ÏµúÏÜå, ÏµúÎåÄ ÌèâÏ†êÏù¥ ÎÇòÌÉÄÎÇ®.
result1
```

##### Mean and Minimum Sell of each category (Í∞Å Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò ÌèâÍ∑† Î∞è ÏµúÏÜå ÌåêÎß§Îüâ)

```python
# 'category_3'Î•º Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôîÌïòÍ≥†, Í∞Å Í∑∏Î£πÏóê ÏÜçÌïòÎäî 'Price' Ïó¥Ïùò ÌèâÍ∑†(mean)Í≥º ÏµúÏÜåÍ∞í(min)ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(result1)ÏùÑ ÏÉùÏÑ±
result1 = df.groupby('category_3')['Price'].agg(['mean', 'min'])

# result1 Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'mean' Ïó¥Ïùò Í∞íÏùÑ ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶ºÌïòÏó¨ Î≥ÄÍ≤Ω
result1['mean'] = result1['mean'].round(2)

# ÏàòÏ†ïÎêú result1 Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï∂úÎ†•
result1

# Ïù¥ ÏΩîÎìúÎäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)ÏùÑ 'category_3' Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôîÌïòÍ≥†,
# Í∞Å Í∑∏Î£πÏóê ÏÜçÌïòÎäî 'Price' Ïó¥Ïùò ÌèâÍ∑†Í≥º ÏµúÏÜåÍ∞íÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(result1)ÏùÑ ÏÉùÏÑ±
# Í∑∏Î¶¨Í≥† ÎÇòÏÑú 'mean' Ïó¥Ïùò Í∞íÏùÑ ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶ºÌïòÏó¨ Î≥ÄÍ≤ΩÌïòÍ≥†,
# ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú ÏàòÏ†ïÎêú result1 Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï∂úÎ†•
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ result1Ïùò Ïù∏Îç±Ïä§Î•º Ïû¨ÏÑ§Ï†ïÌïòÍ≥†, Í∏∞Ï°¥ Ïù∏Îç±Ïä§Î•º Î≤ÑÎ¶¨Î©∞(inplace=True), Ïù¥Î•º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Î∞òÏòÅ
result1.reset_index(inplace=True)

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ result1ÏùÑ 'category_3' Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ¶¨ÌïòÍ≥†,
# 'Stat' Ïó¥ÏóêÎäî 'mean'Í≥º 'min'ÏùÑ, 'Price' Ïó¥ÏóêÎäî Í∞ÅÍ∞ÅÏùò Í∞íÎì§ÏùÑ Ï†ÄÏû•ÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ result_meltedÎ•º ÏÉùÏÑ±
result_melted = result1.melt(id_vars='category_3', var_name='Stat', value_name='Price')

# SeabornÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Í∞Å Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò ÌèâÍ∑† Î∞è ÏµúÏÜå ÌåêÎß§ÎüâÏùÑ ÏÑ† Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôî
# 'category_3'Í∞Ä xÏ∂ïÏóê, 'Price'Í∞Ä yÏ∂ïÏóê, 'Stat'ÏóêÎäî 'mean'Í≥º 'min'ÏùÑ Íµ¨Î∂ÑÌïòÏó¨ Í∞ÅÍ∞ÅÏùò ÏÑ†ÏùÑ ÌëúÏãú
plt.figure(figsize=(10, 6))
sns.lineplot(x='category_3', y='Price', hue='Stat', data=result_melted, marker='o')

# Ï∞®Ìä∏Ïùò Ï†úÎ™©Í≥º ÎùºÎ≤®ÏùÑ ÏÑ§Ï†ï
plt.title('Price Summary by Category')
plt.xlabel('Category')
plt.ylabel('Total Sell')
# 'ÌÜµÍ≥ÑÎüâ'ÏùÄ Î≤îÎ°ÄÏùò Ï†úÎ™©
plt.legend(title='Statistic')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# Í≤∞Í≥ºÎäî Í∞Å Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò ÌèâÍ∑† Î∞è ÏµúÏÜå ÌåêÎß§ÎüâÏùÑ ÎÇòÌÉÄÎÇ¥Îäî ÏÑ† Í∑∏ÎûòÌîÑ
```
